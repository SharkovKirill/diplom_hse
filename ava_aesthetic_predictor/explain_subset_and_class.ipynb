```python
import os
import cv2
import numpy as np
from PIL import Image
import pandas as pd
```

```python
def get_images_by_path(folder_path: str, mode: str):
    images_list = []

    for filename in os.listdir(folder_path):
        if filename.endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(folder_path, filename)
            if mode == "cv2":
                img = cv2.imread(img_path)
            elif mode == "pil":
                img = Image.open(img_path)

            if img is not None:
                images_list.append(img)
            else:
                print(f"Ошибка загрузки: {filename}")

    print(f"Загружено {len(images_list)} изображений.")
    return images_list
```

```python
from ava_aesthetic_predictor.make_prediction_dirt import make_prediction_pretrained_ava

df = []

for subset in ["train", "val", "test"]:
    for estetica_class in range(1, 6):
        print(subset, estetica_class)
        path, mode = (
            rf"D:\MY_PROJECTS\diplom_hse\train_models\prepared_data_2\{subset}\{estetica_class}"
        ), "pil"
        images = get_images_by_path(path, mode)
        scores = np.mean([make_prediction_pretrained_ava(image) for image in images])
        df.append([subset, estetica_class, scores])

final_df = pd.DataFrame(
    df,
    columns=["subset", "estetica_class", "score"],
)
```
